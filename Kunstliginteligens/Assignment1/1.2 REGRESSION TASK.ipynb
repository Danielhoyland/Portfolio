```python
#import statements
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Reading & describing the data 
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Assignement1AI/Height_Weight_Dataset.csv')
df.head()
```

```python
X=df.Age
y=df.Height
X=X.values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression # import the Linear Regression model
lin_reg = LinearRegression() # creat model object
lin_reg.fit(X_train, y_train) # fits the model to the training data
```

```python
plt.scatter(X_train,lin_reg.predict(X_train),color = 'red')
plt.title('predicted height?') # adding a tittle to our plot
plt.xlabel('Age') # adding a label to the x-axis
plt.ylabel('Height') # adding a label to the y-axis
plt.show() # prints our plot
```

```python
from sklearn.preprocessing import PolynomialFeatures # importing a class for Polynomial Regression
poly_regr = PolynomialFeatures(degree = 2) # our polynomial model is of order
X_poly = poly_regr.fit_transform(X_train) # transforms the features to the polynomial form
```

```python
lin_reg_2 = LinearRegression() # creates a linear regression object
lin_reg_2.fit(X_poly, y_train) # fits the linear regression object to the polynomial features
```

```python
plt.scatter(X_train, lin_reg_2.predict(X_poly),color = 'red')
plt.title('predicted height?') # adding a tittle to our plot
plt.xlabel('Age') # adding a label to the x-axis
plt.ylabel('Height') # adding a label to the y-axis
plt.show() # prints our plot
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(lin_reg.predict(X_train), y_train)
error2 = mean_squared_error(lin_reg_2.predict(X_poly), y_train)
print(error)
print(error2)
```
