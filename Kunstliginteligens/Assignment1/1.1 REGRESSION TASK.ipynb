```python
#import statements
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Reading & describing the data 
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Assignement1AI/multivariateregression_data.csv')
df.head()
```

```python
y = df["CO2"].values
X = df.iloc[:,2:4].values
X
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
plt.scatter(X[:, 0], y, color='blue')
plt.scatter(X[:, 1], y, color='green')

# line plot of linear regression model
x_line = np.linspace(np.min(X), np.max(X), 100)
x_line_2d = np.column_stack((x_line, np.zeros_like(x_line)))  # create 2D array with 2 columns
y_line = lin_reg.predict(x_line_2d)
plt.plot(x_line, y_line, color='red')

# add title and labels
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')

# display plot
plt.show()
```
