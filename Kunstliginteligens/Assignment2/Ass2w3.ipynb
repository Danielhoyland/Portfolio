```python
import pandas as pd
from transformers import BertTokenizer, BertModel, TFBertModel
import torch
path_dir = '/content/drive/MyDrive/AIAssignment2/imdb'
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained("bert-base-uncased")
```

```python
df_reviews_train = pd.read_json(path_dir + "/train.json")
```

```python
from transformers import BertTokenizer, BertModel
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained("bert-base-uncased")
text = df_reviews_train.iloc[0]['text']
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)
```

```python
embeddings = [output[0][:,0,:] for input in encoded_input]
embeddings
```

```python
import torch
from sklearn.metrics.pairwise import cosine_similarity

word1 = "comedy"
word2 = "movie"
encoded_inputs = tokenizer([word1, word2], return_tensors='pt', padding=True, truncation=True)

with torch.no_grad():
    outputs = model(**encoded_inputs)
embeddings = outputs.last_hidden_state.squeeze(0)  # Squeeze to remove batch dimension

# Calculate cosine similarity between the word embeddings
similarity = cosine_similarity(embeddings[0].reshape(1, -1), embeddings[1].reshape(1, -1))[0][0]

print(f"Cosine similarity between '{word1}' and '{word2}': {similarity:.4f}")
```
